```python
import plotly.graph_objects as go
import numpy as np
from art import text2art

print_out = text2art("Plotly 3D Surface")


x = y = np.linspace(-3,3,50)
x,y = np.meshgrid(x,y);
Z = np.sin(x**2 + y**2)

fig = go.Figure(data=[go.Surface(z=Z, colorscale='Viridis')])

fig.update_layout(title="3D surface"); 
fig.show()
print(print_out)
```

```python
import plotly.graph_objects as go
import numpy as np
from art import text2art

# Create data grid
x = np.linspace(-3, 3, 50)
y = np.linspace(-3, 3, 50)
x, y = np.meshgrid(x, y)

# Define surface
Z = np.sin(x**2 + y**2)

# Create figure
fig = go.Figure(data=[go.Surface(z=Z, colorscale='Viridis')])

fig.update_layout(
    title="3D Surface Plot using Plotly",
    scene=dict(
        xaxis_title="X Axis",
        yaxis_title="Y Axis",
        zaxis_title="Z Axis"
    )
)

# Show plot
fig.show()
```
