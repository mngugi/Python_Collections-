```python
import tkinter as tk

def update_color():
    r = red_slider.get()
    g = green_slider.get()
    b = blue_slider.get()
       
    # Convert RGB to Hex
    color = f"#{r:02x}{g:02x}{b:02x}"
    
    color_box.config(bg=color)
    hex_label.config(text=f"Hex: {color}")

# Create main window
root = tk.Tk()
root.title("Color Mixer")

# Red Slider
tk.Label(root, text="Red").pack()
red_slider = tk.Scale(root, from_=0, to=255, orient="horizontal", command=lambda x: update_color())
red_slider.pack()

# Green Slider
tk.Label(root, text="Green").pack()
green_slider = tk.Scale(root, from_=0, to=255, orient="horizontal", command=lambda x: update_color())
green_slider.pack()

# Blue Slider
tk.Label(root, text="Blue").pack()
blue_slider = tk.Scale(root, from_=0, to=255, orient="horizontal", command=lambda x: update_color())
blue_slider.pack()

# Color Box
color_box = tk.Label(root, text="Mixed Color", width=20, height=5, bg="white")
color_box.pack()

# Hex Code Label
hex_label = tk.Label(root, text="Hex: #FFFFFF")
hex_label.pack()

# Run the application
root.mainloop()
```

```python
import tkinter as tk

def update_color():
    r = red_slider.get()
    g = green_slider.get()
    b = blue_slider.get()
    w = white_slider.get()
    blk = black_slider.get()

    # Adjust colors based on White and Black influence
    r = min(255, r + w)  # Increase brightness with white
    g = min(255, g + w)
    b = min(255, b + w)

    r = max(0, r - blk)  # Decrease brightness with black
    g = max(0, g - blk)
    b = max(0, b - blk)

    # Convert to Hex
    color = f"#{r:02x}{g:02x}{b:02x}"
    
    # Update UI
    color_box.config(bg=color)
    hex_label.config(text=f"Hex: {color}")

# Create main window
root = tk.Tk()
root.title("Advanced Color Mixer")

# Sliders for Colors
for name, var in [("Red", "red"), ("Green", "green"), ("Blue", "blue"), ("White", "white"), ("Black", "black")]:
    tk.Label(root, text=name).pack()
    globals()[f"{var}_slider"] = tk.Scale(root, from_=0, to=255, orient="horizontal", command=lambda x: update_color())
    globals()[f"{var}_slider"].pack()

# Color Display Box
color_box = tk.Label(root, text="Mixed Color", width=20, height=5, bg="white")
color_box.pack()

# Hex Code Label
hex_label = tk.Label(root, text="Hex: #FFFFFF")
hex_label.pack()

# Run the application
root.mainloop()
```
