```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Function to calculate the angle between two vectors
def angle_between_vectors(v1, v2):
    dot_product = np.dot(v1, v2)
    magnitude_v1 = np.linalg.norm(v1)
    magnitude_v2 = np.linalg.norm(v2)
    cos_theta = dot_product / (magnitude_v1 * magnitude_v2)
    angle_rad = np.arccos(np.clip(cos_theta, -1.0, 1.0))  # Ensures values stay in range
    angle_deg = np.degrees(angle_rad)
    return angle_deg

# Define vectors in 3D space
vector_a = np.array([3, 4, 0])  # Vector A in xy-plane
vector_b = np.array([4, 0, 5])  # Vector B in xz-plane

# Calculate the angle between vectors
angle = angle_between_vectors(vector_a, vector_b)

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the vectors
origin = [0, 0, 0]  # Starting point for vectors
ax.quiver(*origin, *vector_a, color='r', label='Vector A (3, 4, 0)', arrow_length_ratio=0.1)
ax.quiver(*origin, *vector_b, color='b', label='Vector B (4, 0, 5)', arrow_length_ratio=0.1)

# Set labels for axes
ax.set_xlim([0, 5])
ax.set_ylim([0, 5])
ax.set_zlim([0, 5])
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')

# Add title and legend
ax.set_title(f"Angle between Vectors: {angle:.2f}")
ax.legend()

# Display the plot
plt.show()
```
